# 📈 Punjab Economic Census Analysis
### *Insights into the Industrial & Commercial Landscape of Punjab*

**Project Overview:** 
This notebook provides a comprehensive Exploratory Data Analysis (EDA) of the Economic Census data for the Punjab province. We explore the distribution of establishments across districts and identify the key economic sectors driving the workforce.

## 1. Environment Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)
```

## 2. Load and Inspect Data

```python
# Loading the cleaned census data
df = pd.read_csv('punjab_economic_census.csv')
print(f"Data Shape: {df.shape}")
df.head()
```

## 3. District-wise Distribution
Which districts have the highest density of economic establishments?

```python
district_summary = df.groupby('District')[['Establishments', 'Workforce']].sum().sort_values('Establishments', ascending=False)

plt.figure(figsize=(15, 6))
sns.barplot(x=district_summary.index, y=district_summary['Establishments'], palette='magma')
plt.xticks(rotation=90)
plt.title('Total Economic Establishments by District')
plt.show()
```

## 4. Sector-wise Analysis
What are the most common economic activities in Punjab?

```python
sector_summary = df.groupby('Description')[['Establishments']].sum().sort_values('Establishments', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y=sector_summary.index, x=sector_summary['Establishments'], palette='viridis')
plt.title('Top 10 Economic Sectors by Number of Establishments')
plt.show()
```

## 5. Workforce Distribution
Identifying sectors that employ the most people.

```python
workforce_sector = df.groupby('Description')[['Workforce']].sum().sort_values('Workforce', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y=workforce_sector.index, x=workforce_sector['Workforce'], palette='coolwarm')
plt.title('Top 10 Employment-Generating Sectors')
plt.show()
```

## 6. Correlation Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Establishments', y='Workforce', alpha=0.5)
plt.title('Relationship between Number of Establishments and Workforce Size')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

## 7. Professional Insights & Summary
1. **Economic Hubs:** Major districts like Lahore and Faisalabad show the highest concentration of both establishments and workforce.
2. **Primary Sectors:** Retail trade and small-scale manufacturing dominate the count of establishments.
3. **Employment Engines:** While retail has the most units, manufacturing often leads in total workforce per establishment.
4. **Data Quality:** The dataset is clean with minimal missing values, allowing for high-confidence regional planning insights.
