```python
# select top 5 countries by growth in latest year
top5_countries = top10_growth['Country'].head(5).tolist()

# plot growth trends for these countries
plt.figure(figsize=(13, 6))
for country in top5_countries:

    country_data = df_long[df_long['Country'] == country]plt.show()

    plt.plot(country_data['Year'], country_data['GrowthPct'], marker='o', label=country, linewidth=2)plt.tight_layout()

plt.grid(True, alpha=0.3)

plt.title('Population Growth Rate Trends: Top 5 Countries', fontsize=14, fontweight='bold')plt.legend(loc='best')

plt.xlabel('Year', fontsize=12)plt.ylabel('Growth Rate (%)', fontsize=12)
```

## 5. Geospatial EDA 📍

Create an interactive world map showing growth rates by country for the most recent year.

```python
# get latest year data
latest_year = df_long['Year'].max()
df_latest = df_long[df_long['Year'] == latest_year].copy()

print(f"Number of countries with growth data in {latest_year.year}: {len(df_latest)}")
print(f"\nTop 5 countries by growth rate:")
print(df_latest.nlargest(5, 'GrowthPct')[['Country', 'GrowthPct']])
print(f"\nBottom 5 countries (slowest/negative growth):")
print(df_latest.nsmallest(5, 'GrowthPct')[['Country', 'GrowthPct']])
```

```python
# load world boundaries for geospatial visualization
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# simple test: merge and display
world_test = world.copy()
world_test = world_test.merge(df_latest, left_on='name', right_on='Country', how='left')


# create static choroplethplt.show()

fig, ax = plt.subplots(1, 1, figsize=(16, 10))plt.tight_layout()

world_test.plot(column='GrowthPct', cmap='RdYlGn', legend=True, ax=ax,ax.axis('off')

                legend_kwds={'label': f'Growth Rate (%) in {latest_year.year}', 'orientation': 'horizontal', 'pad': 0.05},ax.set_title(f'Population Growth Rate by Country ({latest_year.year})', fontsize=14, fontweight='bold')
                vmin=-5, vmax=5)
```

For interactive mapping, use Plotly's choropleth:

```python
# interactive choropleth with plotly (note: country name matching may require adjustment)
fig = px.choropleth(df_latest, locations='Country',
                    locationmode='country names',
                    color='GrowthPct',
                    hover_name='Country',
                    hover_data={'GrowthPct': ':.2f', 'Country': True},
                    title=f'Population Growth Rate by Country ({latest_year.year})',

                    color_continuous_scale='RdYlGn',fig.show()

                    range_color=(-5, 5))fig.update_geos(showland=True, landcolor='lightgray')
```

## 8. Statistical Insights and Population Effects 📈

Let's compute some basic summary statistics and discuss implications of population growth.

```python
# analyze by continent (if available in world geodataframe)
if 'continent' in world_test.columns:
    continent_growth = world_test.dropna(subset=['GrowthPct']).groupby('continent')['GrowthPct'].agg(['mean', 'count']).sort_values('mean', ascending=False)
    print("\nAverage Growth Rate by Continent:")
    print(continent_growth)
else:
    print("Continent column not available in world dataset.")
```

> **Interpretation:** High average growth rates may indicate young populations or improving healthcare; low/negative rates can signal aging populations or emigration.

We can also calculate world population doubling time using the rule of 70.

```python
# current global statistics
current_rate = global_growth['GrowthPct'].iloc[-1]
print(f"Current global average growth rate: {current_rate:.2f}%")

if current_rate > 0:
    doubling_time = 70 / current_rate
    print(f"Approximate doubling time (Rule of 70): {doubling_time:.1f} years")

# show regional variation
print(f"\nGrowth rate range in {latest_year.year}:")

print(f"  Fastest: {df_latest['GrowthPct'].max():.2f}%")print(f"  Median:  {df_latest['GrowthPct'].median():.2f}%")
print(f"  Slowest: {df_latest['GrowthPct'].min():.2f}%")
```
